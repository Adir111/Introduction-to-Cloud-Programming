```python
import json
from google.colab import drive
import ipywidgets as widgets
drive.mount('/content/drive')
```

```python
# get data
file_path = '/content/drive/My Drive/students.json'
with open(file_path, 'r') as file:
    json_data = file.read()

data = json.loads(json_data)
```

```python
# creates name dictionary
def update_names():
    global names
    names = {
        f"{item['first_name']} {item['last_name']}": item for item in data
    }
update_names()
```

```python
# when name changes in ddl
def on_name_change(change):
    selected_name = change['new']
    if selected_name:
        selected_data = names[selected_name]
        email_text.value = selected_data['email']
        courses_text.value = selected_data['courses']
        link_text.value = selected_data['favorite_show']
```

```python
# define drop down list
name_dropdown = widgets.Dropdown(
    options=list(names.keys()),
    disabled=False,
    layout=widgets.Layout(width='70%')
)

# observe ddl changes
name_dropdown.observe(on_name_change, names='value')
```

```python
# define text fields
email_text = widgets.Text(disabled=True, layout=widgets.Layout(width='70%'), font_family='Tahoma')
courses_text = widgets.Text(disabled=True, layout=widgets.Layout(width='70%'), font_family='Tahoma')
link_text = widgets.Text(disabled=False, layout=widgets.Layout(width='70%'), font_family='Tahoma')
```

```python
# label style
label_layout = widgets.Layout(width='20%', display='flex', justify_content='flex-end', align_items='center', font_family='Tahoma', font_weight='bold')
input_layout = widgets.Layout(width='80%')
```

```python
# make rows for vbox
name_row = widgets.HBox([widgets.Label('Name:', layout=label_layout), name_dropdown], layout=widgets.Layout(margin='5px 0'))
email_row = widgets.HBox([widgets.Label('Email:', layout=label_layout), email_text], layout=widgets.Layout(margin='5px 0'))
courses_row = widgets.HBox([widgets.Label('Courses:', layout=label_layout), courses_text], layout=widgets.Layout(margin='5px 0'))
link_row = widgets.HBox([widgets.Label('Favorite Show:', layout=label_layout), link_text], layout=widgets.Layout(margin='5px 0'))
```

```python
# save data method
def save_data(_):
    selected_name = name_dropdown.value
    if selected_name:
        selected_data = names[selected_name]
        selected_data['favorite_show'] = link_text.value

        # Update the data dictionary directly using selected_name
        data[data.index(selected_data)]['favorite_show'] = link_text.value

        # Write updated data to the file
        with open(file_path, 'w') as file:
            file.write(json.dumps(data, indent=2))

        # Update names dictionary
        update_names()

        save_message.value = '<font color="green">Data saved successfully!</font>'
```

```python
# define save button
save_button = widgets.Button(description='Save')
save_message = widgets.HTML(value='')
save_button.on_click(save_data)
```

```python
# hide save message
def hide_save_message(change):
    save_message.value = ''

link_text.observe(hide_save_message, names='value')
name_dropdown.observe(hide_save_message, names='value')
```

```python
# make form
form = widgets.VBox([name_row, email_row, courses_row, link_row, save_button, save_message], layout=widgets.Layout(width='40%'))
# displays the form
display(form)
# init fields
on_name_change({'new': name_dropdown.value})
```
